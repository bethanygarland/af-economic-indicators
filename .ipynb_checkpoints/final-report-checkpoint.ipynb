### Renaming the columns

```python
af.rename(columns={'Country and Regions': 'Country and Region ID', \
                   'Country and Regions Name': 'Country Name', \
                   'Country and Regions - RegionId': 'Region ID', \
                   'Indicators': 'Indicator ID', 'Indicators Name': 'Indicator Name' }, \
                    inplace=True)
```

```python
af.head(1)
```

### Investigating Nulls

```python
null_df = pd.DataFrame(af.isnull().sum(), columns=['Count of Nulls'])
null_df.index.name = 'Column'
null_df.sort_values(['Count of Nulls'], ascending=False).head(6)
```

Finally, we can see the main culprits of our missing data. A lot of the data is missing from the 1980's. I considered filling the nulls with a mean but since I am interested in time series and how the variable change over time, because there are so many nulls, filling with the mean would not seem useful. I am not focusing on this set of years so I chose to truncate the data and focus on later years with better data.

```python
small_af = af.drop(['1980', '1981', '1982', \
               '1983', '1984', '1985', \
               '1986', '1987', '1988', \
               '1989'], axis=1)
small_af.head(1)
```

Here are the other ways I reduced the nulls after finding them in the data. They are summarised below. I replaced the null in Region ID with 'no ID' because the data type is a string. I removed Eritrea because it had virtually no data but it is important to note this when reviewing the countries later on. I also removed null rows at the bottom. There are more nulls in the data than I have removed but for speed, I chose to remove any big chunks and then remove others when it was necessary during analysis.

```python
small_af.fillna(value={ 'Region ID': 'No ID' }, inplace=True)
aff = small_af.loc[small_af['Country Name'] == 'Eritrea']
new_af = small_af.drop(aff.index, axis=0)
new_af.drop([1731, 1730, 1729],axis=0, inplace = True)
new_af.tail(2)
```

### Setting the Index
My data didn't have an obvious index of it's own, it didn't have any unique identifier for a row so I set one.

```python
new_af['Index'] = new_af.index
new_af.set_index('Index', inplace=True)
```

### Reshaping the data

Because I wanted to be able to compare the data between indicators, I ideally wanted the indicators to be columns and move the year into a column of it's own instead of having every year as a separate column. This would allow me to fill nulls with column averages and get better insight from column descriptive statistics. 

```python
af_melt = new_af.melt(id_vars=["Country and Region ID", \
                               "Country Name", "Region ID", \
                               "Indicator ID", "Indicator Name", \
                               "Scale", "Units"], \
                                var_name="Year", \
                                value_name="Value")
af_melt.head(2)
```

In order to simplify slightly, I chose to try and pivot the data using a particular year. 2018 makes sense 
as it is recent, accurate data that demonstrates Africa's growth.

```python
af_2018 = af_melt[af_melt['Year'] == '2018' ]
af18_pivot = af_2018.pivot_table('Value', ['Year', 'Country Name'], 'Indicator Name')
af18_pivot.head(3)
```

I can now effectively deal with the nulls by filling them instead of dropping them because the values in each column can be used. I will have a look at them first to understand how they are distributed. 

```python
print('Pivot nulls:', af18_pivot.isnull().sum().sum())
null_af = pd.DataFrame(af18_pivot.isnull().sum(), columns=['Count of Nulls'])
null_af.index.name = 'Column'
null_af.sort_values(['Count of Nulls'], ascending=False).head(2)

```

```python
af18_pivot.fillna(af18_pivot.mean(), inplace=True)
print('New Pivot nulls:', af18_pivot.isnull().sum().sum())
```

We will now keep both the af_melt and af2018_pivot data frames in our back pocket ready for analysis.

### Gowth by Region

To start with, I really wanted to get a sense of Africa's growth by region. I decided to use an area plot with Seaborn so I could not only at how each region contributes to growth at an absolute level, but review the varying growth rates of the regions over time. My hypothesis is that the contribution to GDP will be closely linked to the population or geographical size of the area. East Africa has the largest population and of course Sub Saharan Africa covers a lot more countries, so I would expect those areas to be large. I also know that Africa has been growing faster in recent years so I will be interested to see the slope the curves for each region.

```python
af_gdp = af_melt[af_melt['Indicator Name'] == 'Gross domestic product, (constant prices US$)' ]
afgdp_region = af_gdp.loc[af_gdp['Country Name'].isin(['Africa','North Africa', 'East Africa', \
                                                       'Central Africa', 'West Africa', 'Southern Africa', \
                                                       'Sub-Saharan Africa'])]

sns.set_style('darkgrid')
g = sns.FacetGrid(afgdp_region, col='Country Name', hue='Country Name', col_wrap=4 )
g = g.map(plt.plot, 'Year', 'Value')
g = g.map(plt.fill_between, 'Year', 'Value', alpha=0.2).set_titles("{col_name}")

g.set_ylabels("GDP")
g.set_xlabels("Year")
g.set_xticklabels(rotation=45);
```

It is interesting that East Africa has the largest populatation but doesn't seem to contribute as much to overall growth. It is however, growing faster in recent years, the slope is much steeper in the last few years than it has been. This seems to be true of Africa in general and most areas. Clearly Central Africa contributes the least and the speed of the growth doesn't seem to have changed much, which may be due to civil unrest in the area. Sub Saharan Africa looks to be out performing North Africa which is also interesting to note. 

### Growth by country for 2018 and 2019 predictions 

Now I will drill down to try and get a sense of growth on a country by country basis. I expect the top 10 to be very strong figures as they make up some of the fastest growing countries in the world.

### Negative Growth and Instability

Now we have reviewed the fastest growing, it is time to look at the other side of the coin and see which countries are growing more slowly. Some describe the growth performance of the continent as somewhat disappointing, I want to see if I can investigate why that may be. 

```python
low_growth = rate_af18.sort_values(by='2018', ascending = True) \
                      .head(10)
```

I wanted now to find a way to take these countries and find the real GDP growth per capita values for them in order to compare the slow growth in GDP with the adjusted value for population growth. I could have done this arguably more easily using the pivot table, however I find it harder to index into countries with that dataset. What I do, is find the bottom 10 slowest growing countries in the dataset, then use those country names in a list to index into another dataset. This will return the per capita growth rates for the bottom 10 slowest growing countries. I will plot and compare the initial real growth rates with ones adjusted for changes in the population.

```python
list = low_growth['Country Name'].to_numpy()

cap_lowgrowth = small_af18.loc[small_af18['Country Name'].isin(list)] \
[small_af18['Indicator Name'] == 'Real per Capita GDP Growth Rate (annual %)' ]

cap_lowgrowth
stability10 = low_growth.merge(cap_lowgrowth, how='left', on='Country Name', suffixes=('_real','_capita'))
stability10
```

```python
bar_stab = stability10.plot(kind = 'barh', x='Country Name', y= ['2018_real', '2018_capita'], rot=1, \
                                                                      figsize =(12,8)) 
plt.xlabel('Growth rates, real and per capita', fontsize=10)
plt.ylabel('Country', fontsize=10)
plt.yticks(fontsize=8, rotation=30)
plt.title('Bottom 10 Slowest Growing African Countries in 2018')

#bar_stab.invert_xaxis();
```

The blue line showing the, mostly, negative growth rates of real GDP and the orange then showing the real growth per capita, so this accounts for changes in population. The result is very interesting, it shows South Africa and Nigeria, two of Africa's largest economies are pulling down it's average growth rate. Nigeria and South Africa together, make up almost half of the whole continents GDP and both economies are contracting. Even though both countries show a positive GDP growth rate, both are below the rate of population growth so are contracting in real, per capita terms. You would of course expect larger economies to have slower growth rates because they have a much higher starting point, but this is more pertinent than just a slower growth rate. 

### Investigating Ethiopia

I wanted to select a country for further investigation, one that was present in the top fastest growing countries and one that is expected to still be growing in the future. With Libya being a bit of an anomoly, it makes sense to choose Ethiopia. What I am trying to understand, is if the data and the factors that together, make up GDP will give me some insight into the reason why Ethiopia is growing. From my general knowledge, my hypothesis would be to expect export lead growth, or a strongly improving performance on the Current Account to demonstrate an increase in Foreign Direct Investment (FDI) from China and France. Even though this tends to make up a relatively small part of the Current Account, I expect to see growth there. If this were the case and the source of gowth, there is some debate as to whether this growth is beneficial to the people who live there. As with everything, it depends on the nature of that FDI, it is often critisiced for being extractive in nature. For example, FDI resulting from an increase in commodity prices or mining and extracting elements from the soil, where the employees of the companies undertaking these activities are not local citizens, can be considered extractive. FDI that flows in and is more long term, sustainable investment in infrastructure is much more beneficial to local communities as it results in higher employment and those employees spending with local businesses and so on. 

```python
year_col = pd.to_datetime(af_melt['Year'])
af_melt['Year'] = year_col 
Eth = af_melt[(af_melt['Country Name'] == 'Ethiopia') & (af_melt['Year'] > '2010-01-01')]
Eth_indicators = Eth.loc[Eth['Indicator Name'].isin(['Final consumption expenditure (current US$)', 'Gross capital formation (current US$)','General government final consumption expenditure (current US$)', 'Gross capital formation (current US$)', 'Gross capital formation, Private sector  (current US$)', 'Gross capital formation, Public sector  (current US$)', 'Exports of goods and services (current US$)', 'Imports of goods and services (current US$)', 'Central government, Fiscal Balance (Current US $)', 'Current account balance (Net, BoP, cur. US$)'])]
e = sns.FacetGrid(Eth_indicators, col='Indicator Name', hue='Indicator Name', col_wrap=4, height =5, sharey=True, aspect=1.7 )
e = e.map(plt.plot, 'Year', 'Value')
e.set_ylabels("Indicator Value")
e.set_xlabels("Year")

sns.set(font_scale=0.9)
```

The results here are somewhat different to what I hypothesised. Consumption is steadily increasing which is to be expected but it is not climbing rapidly and is not contributing the most heavily to growth (it often is the largest component). If we take a look at the Current Account, which is where any FDI would be present, it is actually dropping slightly. Exports are also rising but very slowly, so although export growth is here, it is not the largest factor responsible for growth. The Current Account may be decreasing due to the increase of imports of goods and services. This is actually very common during periods of growth in countries, when incomes rise, people's propensity to purchase imports also rises. Higher income tends to lead to buying more imported goods, so net-net this may be why the current account isn't improving dramatically. One of the most pertinent aspects to note here, is the sharp rise in Gross Capital Formation. This is another way to describe the countries investments, the second eleemt that makes up total GDP for a given country. The value of Investment also, is most noticably rising in the Public Sector, so Government spending on investment projects which may also account for the rise in the level of Government spending. After some research, it looks to be correct that Ethiopia is growing speedily due to investment in infrastructure and manufacturing. They are currently building the Grand Renaissance Damn which will be the largest on the continent. They are also investing in other infrastructure like roads and rail links which will prove invaluable to future growth by connecting villages and towns and allowing businesses to travel across the country. This is much more promising for long term growth than it being propped up by rising commodity prices or exports which are vulnerable to external shocks. It often benefits local people more, as it less extractive in nature, particularly when it comes to employment. With rising employment from the government in these regions, people have 'more money in their pocket' which they spend on local businesses and also may attract more businesses and even global retailers. With more income, it is possible to buy better food and supplies, improve housing, healthcare and education. It boosts the economy exponentially. It is hard to argue that lifting a large number of people out of poverty would be a bad thing but growth in developing countries can often lead to inflation, so the benefit will always depend on the effect and distribution. It also can have a detrimental effect on local culture. We tend to think in our own more western terms of growth being good and boosting employment etc. but a lot of people don't want to attract the likes of Starbucks and H&M and so fourth. They want to preserve culture and not have destroyed by this rapid desire for 'growth at all costs'. The debate is a fascinating one, however it bodes well that the growth is investment lead because it will pay dividends for years to come and could pave the way for a real economic powerhouse.

```python
af18_countries = af18_pivot[af18_pivot['Gross domestic product, current prices (current US$)'] <= 500000]
af18_countries.plot(kind='scatter', x='Gross domestic product, current prices (current US$)', \
                                    y='Inflation, consumer prices (annual %)', figsize=(10,8), \
                                    s=10, color='darkblue' )

plt.title('The relationship between GDP and Inflation in Africa 2018');
```

The relationship between growth or employment and prices/ inflation is one of the most researched relationships in economics, so it's an important one to investigate. Ideally you wouldn't have Inflation as a percentage here but rather prices, so the units and scale is difficult to measure and show. Even using Inflation as a percentage, there is clearly a relationship here. As we know, when Aggregate Demand in an economy expands faster than aggregate supply, we expect to see a higher inflation rate. This can be due to supply constraints, firms can't keep up with demand so they raise prices or more of a Phillips Curve type relationship where it is due to wage inflation. Perhaps what we have here is somewhat of a Phillips Curve, although he uses wage inflation and unemployment. The idea being when unemployment is low, there is scarcity of labour and so employers will raise wages to attract scarce labour. This later became general inflation because of the relationship between how much a company pays in wages and how much it charges for it's goods and services. This approach or theory was later critisised fairly famously by Milton Friedman. Again, this may be an argument for another report but it is interesting to visualise the relationship here. Again, I am reluctant to use a thoery that was developed in the UK to apply to countries in Africa because there are so many different factors at play. African prices are often sensitive to food prices so can be affected by weather conditions or bad harvests. They also fall victim to imported inflation resultant from currency depreciation and fluctuations in commodity markets. It is more common in Africa to experience a currency crises, so there are many reasons why it is hard to assume a UK growth-inflation relationship. Also, every country in Africa is so different and should have many considerations, in that way you could argue there is no such thing as 'African Inflation'(Hans Frances, Janssens, 2018). Having said that, it is hard to argue with the notable observation that inflation in general is very high. For reference the US has a fairly steady 1.7% inflation rate. What is also interesting is that some of the larger economies, to the right of the x axis, have managed to sustain growth at lower inflation. Nigeria being one of them. Perhaps it is not all disappointment for the larger economies, even thought they are growing more slowly, they manage to keep their inflation rate relatively low compared to some other countries in the continent.

South Sudan has a staggeringly high inflation rate. Reuters offers the following insight on this:

```python
scatt = af18_countries.plot(kind='scatter', x='Current account balance (Net, BoP, cur. US$)', \
                                            y='Central government, Fiscal Balance (Current US $)', \
                                            figsize=(10,8), s=10, color='darkblue' )
plt.title('The relationship between Government Fiscal Balance and the Current Account in Africa 2018')
scatt.invert_yaxis()
scatt.invert_xaxis();
```

This is a relationship uncovered during Pandas Profiling which I certainly didn't hypothesis or expect. It is a relationship between the Government's Fiscal balance, it's spending and revenue, and the Current Account which is a measure of net exports. 

Some economists believe a large budget deficit can be correlated with a large current account deficit. What we may have found here is sometimes known as the 'twin deficit hypothesis'. The logic behind the theory is that government tax cuts, which reduce revenue and increase the deficit, result in increased consumption as taxpayers spend their new-found money. The increased spending reduces the national savings rate, causing the nation to increase the amount it borrows from abroad. Perhaps this chart is demonstrating that relationship, but it would require more research and analysis over time.
